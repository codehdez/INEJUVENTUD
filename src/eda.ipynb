Filtrar columnas a utilizar|


```python
dfa = df.iloc[:, [1] + list(range(15, 25))]
```

```python
dfa.head()
```

Vamos a sumar la culumna PADRON_18_HOMBRES + PADRON_18_MUJERES + PADRON_18_NOBINARIO 

```python
dfa['suma_total'] = dfa['PADRON_18_HOMBRES'] + dfa['PADRON_18_MUJERES'] + dfa['PADRON_18_NOBINARIO']
```

```python
dfa.head()
```

```python
resultado_agrupado = dfa.groupby('NOMBRE\nENTIDAD' ).sum().sort_values(by='suma_total', ascending=False)
```

```python
resultado_agrupado
```

Renombramos la columnas suma_total

```python
columnas_seleccionadas = resultado_agrupado[['suma_total']]
```
