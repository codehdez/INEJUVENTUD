## Distribución Demográfica por estado de Ciudadanos de 18 a 29 Años en Orden Descendente

 Importamos librerias

```python
import pandas as pd
import numpy as np
```

Importamos los datos            

```python
df = pd.read_excel('../data/ine.xlsx', header=0 )
```

Informacion general de dataframe

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Nombre de todas las comunas que tiene el dataframe

```python
nombres_columnas = df.columns
print(nombres_columnas)
```

Filtrar columnas a utilizar|


```python
dfa = df.iloc[:, [1] + list(range(51, 63))]
```

```python
dfa.head()
```

Vamos a sumar la culumna PADRON_18_HOMBRES + PADRON_18_MUJERES + PADRON_18_NOBINARIO 

```python
dfa['suma_total'] = dfa.iloc[:, 1:27].sum(axis=1)
```

```python
dfa.head()
```

Realizamos la agrupacion por entidad

```python
resultado_agrupado = dfa.groupby('NOMBRE\nENTIDAD').sum().sort_values(by='suma_total', ascending=False)
```

```python
resultado_agrupado
```

Agrupamos por variable suma total

```python
resultado_agrupado = resultado_agrupado['suma_total']
```

```python
resultado_agrupado
```
