Vamos a sumar a sumar todas las columnas a excepcion de la columna 0 y a su vez agregaremos una nueva columna llamada suma_total

```python
dfa['suma_total'] = dfa.iloc[:, 1:27].sum(axis=1)
```

```python
dfa
```

Quitamos espacios de las columnas

Ahora vamos a grupar por estado

```python
resultado_agrupado = dfa.groupby('NOMBRE\nENTIDAD').sum().sort_values(by='suma_total', ascending=False)
```

```python
resultado_agrupado1 = resultado_agrupado.loc[:, ['suma_total']]
```
